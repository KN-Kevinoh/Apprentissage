```python
import pandas as pd
import numpy as np
```

```python
path_dataset = "dat/"
```

```python
file = open(path_dataset + "features.txt", "r")
lines = file.readlines()
file.close()

features = []

for line in lines:
    features.append(line.split(' ')[1])
    
features.append('id_person')
features.append('label')
    
# Création du tableau de données
dataset = pd.DataFrame(None, columns=features)
```

```python
dataset
```

```python
file_train = open(path_dataset + "train/X_train.txt", "r")
file_subject = open(path_dataset + "train/subject_train.txt", "r")
file_y = open(path_dataset + "train/y_train.txt", "r")

lines_train = file_train.readlines()
file_train.close()

lines_subject = file_subject.readlines()
file_subject.close()

lines_y = file_y.readlines()
file_y.close()

for i, line in enumerate(lines_train):
    line = line.strip() # On supprime les espaces avans et après la ligne
    line = line.replace("  ", " ") # On supprimer les doubles espaces
    data = line.split(' ') # On recupère chaque valeur
    for j, d in enumerate(data): # On change toutes les valeur en réelle
        data[j] = float(data[j])
    # On ajoute l'id de la personne et le label
    data.append(int(lines_subject[i]))
    data.append(int(lines_y[i]))
    df = pd.DataFrame(data=[data], columns=features)
    dataset = dataset.append(df, ignore_index = True)
```

```python
# Sauvegarde du fichier
dataset.to_csv(r'{}train/X_train_new.txt'.format(path_dataset), header=None, index=None, sep=' ', mode='a')
```
