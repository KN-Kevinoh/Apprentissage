```julia
#lecture d'un fichier
using DelimitedFiles
using Plots
```

```julia
A=readdlm("data/forbes.txt", ' ')
```

```julia
temperature=A[1,:]# y
pressure=A[2,:];#X  -- on ajoute ';' à la fin pour ne pas afficher le resultat
```

```julia
plot(temperature,pressure,st=:scatter, color=:red,xlabel=:"Temperature", ylabel=:Pressure, label=:"",markersize=:5)
#st=serie_time
```

```julia
#y est la temperature et x la pressure
y=pressure #on recupère les valeur de pressure
N=length(y) #la longuer de y
un=ones(N) #on crée un vecteur N 1 pour ajouter a la matrice X
X=[un temperature];#pour ajouter le 1 à la matrice X pour pouvoir calculer theta
```

```julia
ones(N);
```

```julia
θ=inv(X'*X)*X'*y
```

```julia
y_207=[1 207]*θ
```

```julia
plot!([207],y_207,st=:scatter, color=:blue,label=:"",markersize=:8)#on plote le noveau point prédit
```

```julia
ŷ=X*θ
```

```julia
u=ŷ-y
u'*u
```

```julia
#pour ploter la ligne
plot!([temperature[1];temperature[N]],[ŷ[1];ŷ[N]], label=:"",color=:"green",linewidth=3)
#
```
