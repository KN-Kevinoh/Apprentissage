```julia
#lecture d'un fichier
using DelimitedFiles
M=readdlm("data/wdbc.txt", ',')
```

```julia
y=Int.(M[:,2] .+1)#extraction de la colone 2
```

```julia
X=M[:,3:12] #extraction des 10 première features
```

```julia
function train(X,Y)
    N,D=size(X)
    K=length(unique(Y))
    θ=zeros(K)
    μ=zeros(K,D)
    σ=zeros(K,D)
    for k=1:K
        θ[k]=sum(Y.==k)/N
        X_k=X[y.==k,:]
        μ[k,:]=mean(X_k,dims=1)
        σ[k,:]=std(X_k,dims=1)
    end
    (θ,μ,σ)#retour des valeurs
end
```

```julia
θ,μ,σ=train(X,y)
```

```julia
function classify(x,θ,μ,σ)
    K=length(θ)
    posterior=zeros(K)
    for k=1:K
        u=(x-μ[k,:])./σ[k,:]
        posterior[k]=-sum(log.(σ[k,:]))-u'*u/2+log(θ[k])
    end
    argmax(posterior)
end
```

```julia
function appel_prediction(x_i)
    prediction=classify(X[x_i,:],θ,μ,σ)
    if(prediction==1)
        print("1-Cancer")
    else
        print("2-Non-cancer")
    end
end
```

```julia
print("Entrez l'indice du patient:")
x_i= parse(Int, readline())
appel_prediction(x_i)
```

```julia
#évaluation du model
function evaluate(X,y,θ,μ,σ)
    N=length(y)
    ŷ=zeros(N)
    for i=1:N
        ŷ[i]=classify(X[i,:],θ,μ,σ)
    end
    sum(ŷ.==y)/N
end
    
```

```julia
evaluate(X,y,θ,μ,σ)
```
