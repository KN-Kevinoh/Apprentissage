```julia
using Plots
```

```julia
θ=-3.0:0.1:3
```

```julia
J(z)=2*z*z - 3*z+1
```

```julia
y=J.(θ);
```

```julia
plot(θ,y)
```

```julia
#calcul du meuilleur
θ_best=3.0/4
```

```julia
J_best=J(θ_best)
```

```julia
plot!([θ_best],[J_best],st=:scatter, color=:red,label=:"Best")
```
